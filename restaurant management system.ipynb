```python
df= pd.read_csv('Used_Bikes.csv')
df.head
```

```python
df.drop_duplicates(inplace=True)
```

```python
brand_name = list(df['brand'].value_counts().keys())
bike_count = list(df['brand'].value_counts().values)
```

```python
#matplot
plt.bar(brand_name,bike_count)
plt.show()
```

```python
# seaborn
sns.barplot(x=brand_name,y=bike_count)
plt.show()
```

```python
df = sns.load_dataset('tips')
df.head
```

```python
sns.relplot(x='tip',y='total_bill',data=df)
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker')
plt.show()
```

```python
df
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='time',style='smoker',col='day')
plt.show()
```

```python
sns.relplot(x='tip',y='total_bill',data=df,hue='sex',col='day',kind='scatter')
plt.show()
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
sns.relplot(x='size',y='total_bill',data=df,col='day',kind='line')
plt.show()
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex')
plt.show()
# strip chart
```

```python
sns.catplot(x='smoker',y='total_bill',data=df,hue='sex',kind='swarm')
plt.title('swarm chart')
plt.show()
```
